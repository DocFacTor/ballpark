## Summary

This paper introduces heterogeneous-agent macroeconomic model into canonical open economy macroeconomic model.

The main findings of the paper are as follows:

* A HA model with uninsurable aggregate and idiosyncratic risk exhibits larger effect of a real exchange rate depreciation via two channels: real income channel and multiplier channel.

* With low trade elasticity, the two channels can dominate the consumption switching channel, leading to contractionary depreciation.

* The model shows contrasting monetary policy response to a foreign exchange rate hike in the emerging economies and the advanced economies. 

We will refer to the goods of the home (Brazil) as Brazil Nuts (nuts) and that of the foreign (US) as Ford cars (cars).

## The model

The model builds on the canonical open-economy model of Gali and Monacelli (2005) with some additional ingredients:
incomplete markets, heterogeneous households and sticky wages as in Aucert, Rognlie and Straub (2018)

#### Domestic (Brazil) households

The economy is pouplated by a continuum of households. 
Each household is subject to idiosyncratic shock to its productivity ($e$).
This risk cannot be insured.

A household with asset level $a$ and productivity level $e$ at time t chooses her consumption of home (let's say Brazil) and foreign goods as well as her saving, by solving following dynamic program.

$$
\begin{aligned}
V_t(a, e)= & \max _{c_F, c_H, a^{\prime}} u\left(c_F, c_H\right)-v\left(N_t\right)+\beta \mathbb{E}_t\left[V_{t+1}\left(a^{\prime}, e^{\prime}\right)\right] \\
\text { s.t. } & \frac{P_{Ft}}{P_t} c_F+\frac{P_{H t}}{P_t} c_H+a^{\prime}=\left(1+r_t^p\right) a+e \frac{W_t}{P_t} N_t \\
& a^{\prime} \geq \underline{a}
\end{aligned}
$$

where per period utility functions are defined as follows:

$$
u\left(c_F, c_H\right)=\frac{c^{1-\sigma}}{1-\sigma}, \quad v(N)=\psi \frac{N^{1+\varphi}}{1+\varphi}
$$

with $c$ the consumption basket

$$
c=\left[\alpha^{1 / \eta} c_F^{(\eta-1) / \eta}+(1-\alpha)^{1 / \eta} c_H^{(\eta-1) / \eta}\right]^{\eta /(\eta-1)}
$$

The consumer price index ($P_t$) is defined as follows: 

$$
P_t \equiv\left[\alpha P_{F_t}^{1-\eta}+(1-\alpha) P_{H_t}^{1-\eta}\right]^{1 /(1-\eta)}
$$

A household in state $(a,e)$ splits her purchases between foreign and home (Brazilian) goods according to

$$
\begin{aligned}
c_{Ft}(a, e) & =\alpha\left(\frac{P_{F_t}}{P_t}\right)^{-\eta} c_t(a, e) \\
c_{Ht}(a, e) & =(1-\alpha)\left(\frac{P_{H_t}}{P_t}\right)^{-\eta} c_t(a, e)
\end{aligned}
$$

Monetary policy abroad keeps the price of foreign (US) goods in foreign currency (dollar) constant.
(i.e. Jerome Powell keeps the price of Ford fixed.)
We assume that imports are denominated in foreign currency and therefore there is perfect pass-through of exchange rates into domestic goods price. i.e. $P_{F_t}=\mathcal{E}_t$ where $\mathcal{E}_t$ is the nominal exchange rate.

The ***real exchange rate*** is then given by

$$
Q_t \equiv \frac{\mathcal{E}_t}{P_t}
$$

#### Foreign (US) households's consumption of home goods (nuts).

A foreign household has the elasticity of substitution between Home goods (nuts) and Foreign goods (cars) of $\gamma$.
Denoting the foreign-currency price of home (Brazil) produced goods (nuts) by $P_{H_t}^*$, export demand for home (Brazilian) goods is given by

$$
C_{H_t}^* = \alpha \left(\frac{P_{H_t}^*}{P^*}\right)^{-\gamma}C^*
$$

That is, $\eta$ is the elasticity that Brazilians have between nuts and cars and $\gamma$ is the elasticity that Americans have between nuts and cars.

However, households in both countries have the same level of home bias.

#### Trade elasticities of imports and exports.

The elasticity of imports with respect to the relative price of imports $P_{F_t}/P_{H_t}$ is $\eta(1-\alpha)$.
And the foreign counterpart of it is $\gamma$.
We denote $\textit{trade elasticity } \chi$ as the sum of these two. That is:

$$
\chi \equiv \eta(1-\alpha)+\gamma
$$

The latter term for foreign demand of home (Brazilian) good does not depend on foreign home bias because the home country is too small to affect the foreign CPI.

Y_t = Z N_t

The Brazilian firm returns following real dividend.
$$
D_t = \frac{P_{H_{t}}Y_t-W_tN_t}{P_t}+\frac{\mathcal{E}_t P_{H_{t}}^*-P_{H_{t}}}{P_t}C_{H_{t}}^*
$$

Firms have a unit mass of shares outstanding whose end-of-period price is denoted $p_t$.
A firm is a maximizer of $D_t+p_t$.

$N_t$ is aggregate labor supplied and $Z$ is the constant level of labor productivity.
A continuum of monopolistically competitive firms produce home goods with elasticity of substitution between goods from various firms $\epsilon$.

Prices are assumed to be fully flexible for now, therefore the price of home goods is set at a constant markup $\mu$ as follows:

$$
P_{H_{t}} = \mu \frac{W_t}{Z}
$$
where $\mu=\epsilon/(\epsilon -1)$.

#### Financial sector

We assume frictionless capital flows across countries.
An unconstrained, risk-neutral mutual fund issues claims to households, with aggregate real value of $A_t$ at the end of period $t$.

We define return on Brazilian bonds in *Real* $i_t$ and return on Americal bonds in dollars $i_t^{*}$.

Expected returns from different assets are equal, resulting in following real UIP condition.

$$
1+i_t = (1+i_t^*)\frac{Q_{t+1}}{Q_t}
$$

We define the ex-ante real interest rate as
$$
1+r_t \equiv (1+i_t) \frac{P_t}{P_{t+1}}
$$

Because the mutual fund is risk-neutral,

\begin{align}
1+r_t &= \frac{p_{t+1}+D_{t+1}}{p_t} \\
&= \frac{(p_{t+1}+p_{t}-p_{t})+D_{t+1}}{p_t}\\
&= \frac{\overbrace{(p_{t+1}-p_{t})}^{\text{Capital gain}}+D_{t+1}}{p_t}+1
\end{align}


We define net foreign asset position to be

$$
\text{nfa}_t \equiv A_t - p_t
$$

#### Unions

We follow a formulation for sticky wages with heterogeneous households similar to Auclert, Rognlie and Straub (2018).

A union employs all households for an equal number of hours $N_t$ and sets nominal wages at a level that maximizes the welfare of the average household.
We choose the union objective function that leads to the wage Phillips curve:

$$
\pi_{w_t} = \kappa_w \left( \frac{v'(N_t)}{\frac{1}{\mu} \frac{W_t}{P_t} u'(C_t)}-1 \right)+\beta \pi_{w_{t+1}}
$$

where $\pi_{w_t}$ demnotes nominal wage inflation.

#### Monetary policy

The monetary authority sets the nominal interest rate according to a monetary rule.
To get analytical results, we consider a constant real rate rule with shock:

$$i_t = r_{ss}+\pi_{t+1}+\upsilon_t$$


## Exchange rate shocks

We consider shocks to foreign interest rates $i_t^*$, caused by a foreign preference shock $\mathcal{B}_t = \Pi_{s\geq t}\left( \frac{1+i_s^*}{1+r_{ss}} \right)$

Combining the real UIP condition, the long run real exchange rate ($Q_{\infty}=1$), and the monetary rule, we find that the real exchange rate is given by

$$
Q_t = \Pi_{s\geq t}\left( \frac{1+i_s^*}{1+r_{ss}} \right) = \mathcal{B}_t
$$

The analysis is centered around **the home good market clearing condition**, which can be written as follows under Producer Currency Pricing.

$$
Y_t=(1-\alpha)\left(\frac{P_{H_t}}{P_t}\right)^{-\eta} C_t+\alpha\left(\frac{P_{H_t}}{\mathcal{E}_t}\right)^{-\gamma} C^*_t
$$

We contrast a case of complete market with a representative agent against an incomplete market with heterogeneous agents.

### Complete-market, representative agent model 

The exposition of this specification can be found in Gali and Monacelli (2005).

In the complete-market representative agent model where the interest rate follows
$i_t = r_{ss}+\pi_{t+1}+\upsilon_t$,
the linearized deviations of consumption from steady state consumption over output $dC_t$ and output $dY_t$ in response to the
real exchange rate's deviation $dQ_t$ are given by

$$
\begin{align}
    dC_t & = 0 \quad \forall t \\
    dY_t & = \frac{\alpha}{1-\alpha}\chi dQ_t \quad \forall t
\end{align}
$$

### Incomplete market, heterogeneous agent model

Combining the price setting condition and the production function, households' real wage income is given by

$$
\frac{W_t}{P_t}N_t = \frac{1}{\mu} \frac{P_{H_{t}}}{P_t}Y_t
$$

Then real dividends are equal to

$$
D_t = \left( 1-\frac{1}{\mu} \right) \frac{P_{H_{t}}}{P_t}Y_t
$$

Dividend determines households' capital income through real interest rate.
Therefore, the path of aggregate consumption $C_t$ is a function of the path of aggregate real income $\frac{P_{H_{t}}}{P_t}Y_t$.
We denote this "consumption function" by 
$$
C_t = \mathrm{C}_t \left( \left[ \frac{P_{H_{s}}}{P_s}Y_s \right]_{s=0}^{\infty} \right) 
$$


#### Proposition
In response to a shock to the real exchange rate $d\mathbf{Q}$, the impulse response of consumption and output is given by

$$
d \mathbf{C}=-\underbrace{\frac{\alpha}{1-\alpha} \mathbf{M} d \mathbf{Q}}_{\text {Real income channel }}+\underbrace{\mathbf{M} d \mathbf{Y}}_{\text {Multiplier }}
$$

$$
d \mathbf{Y}=\underbrace{\frac{\alpha}{1-\alpha} \chi d \mathbf{Q}}_{\text {Exp. switching channel }}-\underbrace{\alpha \mathbf{M} d \mathbf{Q}}_{\text {Real income channel }}+\underbrace{(1-\alpha) \mathbf{M} d \mathbf{Y}}_{\text {Multiplier }}
$$

$\mathbf{M}$ denote the derivative of $\mathcal{C}$, whose elements $M_{t,s}\equiv \frac{\partial \mathcal{C}_t}{\partial Y_s}$

Rearranging the terms, we get

$$
d \mathbf{Y}=\underbrace{\left(\sum_{k \geq 0}(1-\alpha)^k \mathbf{M}^k\right)}_{=(\mathbf{I}-(1-\alpha) \mathbf{M})^{-1}}\left(\frac{\alpha}{1-\alpha} \chi d \mathbf{Q}-\alpha \mathbf{M} d \mathbf{Q}\right)
$$

## Contractionary depreciations

When $\chi$ is sufficiently below 1, a depreciation of domestic currency leads to contraction in output.

#### Proposition
If $x<1-\alpha$, the output response to a depreciation shock $d Q_t \geq 0$ has negative net present value, $\sum_{t=0}^{\infty}(1+r)^{-t} d Y_t<0$ in the heterogeneous-agent model. Moreover, given a depreciation shock, there is a threshold $\chi^*$ between $(1-\alpha) M_{0,0}$ and 1 such that for any $\chi<\chi^*$, the output response is negative on impact, $d Y_0<0$.

### Monetary policy

A monetary policy has two distinct effects: a change in the interest rate alters the path of the real exchange rate, and affects asset prices and their return.
Hence, monetary policy affects output through four distnct channel: interest rate channel, real income channel, multiplier, and expenditure switching channel.

An accommodative shock in the interest rate has lower or higher impact on output depending on value of trade elasticity.
Following proposition elaborates this result.

#### Proposition 
Assume $\sigma=1$, and consider an arbitrary first-order monetary policy shock $d\mathbf{r}$. If $\chi=2-\alpha$, all aggregate quantities and prices are identical in heterogeneous and representativeagent models. Moreover, provided that $\mathbf{M}>0$, for an accommodative shock $d \mathbf{r} \leq 0$,
$$
d \mathbf{Y} \lessgtr d \mathbf{Y}^{R A} \quad \Leftrightarrow \quad \chi \lessgtr 2-\alpha
$$
